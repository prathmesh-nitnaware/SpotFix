```python
import pandas as pd
import random

# Core Architectural Scenarios
scenarios = {
    "High": {
        "templates": [
            "Emergency: {} at {}", "Critical {} in the {}", "Immediate attention required: {} - {}",
            "Dangerous {} near {}", "{} detected at {} - SOS", "{} is causing a hazard in {}"
        ],
        "issues": ["Fire outbreak", "Live sparking wires", "Severe flooding", "Gas leak", "Short circuit", "Structural collapse", "Stuck elevator", "Broken chemical bottle", "Exposed high-voltage cable"],
        "locations": ["Lab 402", "Main Auditorium", "Canteen", "Hostel Block A", "Chemistry Wing", "Exam Hall", "Server Room"]
    },
    "Medium": {
        "templates": [
            "Broken {} in {}", "The {} is not working properly at {}", "{} needs repair in {}", 
            "{} is leaking in the {}", "Malfunctioning {} at {}", "{} is damaged in {}"
        ],
        "issues": ["Air conditioner", "Projector", "Ceiling fan", "Water tap", "Window pane", "Door lock", "WiFi Router", "Water cooler", "Laboratory equipment"],
        "locations": ["Classroom 101", "Library", "Staff Room", "Washroom", "Gymkhana", "Placement Cell", "Seminar Hall"]
    },
    "Low": {
        "templates": [
            "Need extra {} for {}", "Requesting {} in {}", "The {} is slightly loose at {}", 
            "Please clean the {} in {}", "{} is flickering in {}", "Minor {} issue at {}"
        ],
        "issues": ["Chairs", "Dustbins", "Tube light", "Door handle", "Whiteboard marker", "Notice board", "Signage", "Curtains"],
        "locations": ["Corridor", "Parking Lot", "Garden Area", "Reception", "Waiting Lounge", "Study Area"]
    }
}

dataset = []
categories = ["Electrical", "Plumbing", "IT/AV", "HVAC", "Furniture", "Safety"]

# Generate 1000 Entries
while len(dataset) < 1000:
    for priority, content in scenarios.items():
        if len(dataset) >= 1000: break
        
        template = random.choice(content["templates"])
        issue = random.choice(content["issues"])
        location = random.choice(content["locations"])
        
        # Build the final description
        description = template.format(issue, location)
        category = random.choice(categories)
        
        dataset.append({
            "description": description,
            "priority": priority,
            "category": category
        })

df = pd.DataFrame(dataset)
# Shuffle to ensure random distribution during training
df = df.sample(frac=1).reset_index(drop=True)
df.to_csv('models/advanced_priority_data.csv', index=False)

print(f"Generated {len(df)} advanced entries in backend-python/models/advanced_priority_data.csv")
```

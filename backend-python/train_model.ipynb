```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import GradientBoostingClassifier # Advanced choice over RF
import joblib

# Load the 1000 entries
df = pd.read_csv('models/advanced_priority_data.csv')

# Vectorization
vectorizer = TfidfVectorizer(ngram_range=(1, 2), stop_words='english')
X = vectorizer.fit_transform(df['description'])
y = df['priority']

# Training an Advanced Gradient Boosting model
model = GradientBoostingClassifier(n_estimators=200, learning_rate=0.1)
model.fit(X, y)

# Save both components to your models folder
joblib.dump(model, 'models/priority_model.pkl')
joblib.dump(vectorizer, 'models/vectorizer.pkl')

print("Advanced Intelligence Trained & Saved.")
```
